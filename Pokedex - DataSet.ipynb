```python
for search_term2 in pokemon_list:
    # build the path to the output image
    dirout2=dirout+search_term2
    os.mkdir(dirout2)
    os.chmod(dirout2,0o777)
    #headers and search parameters
    headers = {"Ocp-Apim-Subscription-Key" : subscription_key}
    params  = {"q": search_term2, "license": "Any", "imageType": "photo","count": group_size}
    # get the results and make the search
    response = requests.get(search_url, headers=headers, params=params)
    response.raise_for_status()
    results = response.json()
    #download the images
    i=0
    for v in results["value"]:
        # build the path to the output image
        try:
            ext = v["contentUrl"][v["contentUrl"].rfind("."):]
            r = requests.get(v["contentUrl"], timeout=30)
            f = open(dirout2+'\\'+str(i)+'.png', "wb")
            f.write(r.content)
            f.close()
            i=i+1
        except:
            print("Error")
```
